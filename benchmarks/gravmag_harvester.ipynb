```python
%matplotlib inline
import numpy as np
from fatiando import gridder, utils
from fatiando.gravmag import harvester, prism
from fatiando.gravmag.blocky import Gravity
from fatiando.mesher import Prism, PrismMesh, vremove
from fatiando.vis import mpl, myv
```

```python
from line_profiler import LineProfiler
```

```python
model = [Prism(300, 700, 300, 700, 200, 600, {'density':1000})]
shape = (50, 50)
bounds = [0, 1000, 0, 1000, 0, 1000]
area = bounds[0:4]
xp, yp, zp = gridder.regular(area, shape, z=-1)
noise = 0.1 # 0.1 mGal noise
gz = utils.contaminate(prism.gz(xp, yp, zp, model), 0.1, seed=0)
gzz = utils.contaminate(prism.gzz(xp, yp, zp, model), 5, seed=0)
mesh = PrismMesh(bounds, (10, 10, 10))
#data = [harvester.Gz(xp, yp, zp, gz), harvester.Gzz(xp, yp, zp, gzz)]
data = [harvester.Gz(xp, yp, zp, gz)]
seeds = harvester.sow([[500, 500, 450, {'density':1000}]], mesh)
compactness, threshold = 10, 0.0001
solver = Gravity(xp, yp, zp, gz, mesh) #+ Gravity(xp, yp, zp, gzz, mesh, field='gzz')
args = dict(method='planting', seeds=seeds, compactness=compactness, threshold=threshold)
solver.config(**args)
```

```python
estimate, predicted = harvester.harvest(data, seeds, mesh, compactness, threshold)
mesh.addprop('density', estimate['density'])

mpl.figure(figsize=(14, 5))
#mpl.subplot(1, 2, 1)
mpl.title("gz")
mpl.axis('scaled')
levels = mpl.contourf(yp, xp, gz, shape, 12)
mpl.colorbar()
mpl.contour(yp, xp, predicted[0], shape, levels, color='k')
mpl.xlabel('Horizontal coordinate y (km)')
mpl.ylabel('Horizontal coordinate x (km)')
mpl.m2km()

mpl.figure()
residuals = gz - predicted[0]
mpl.title('Residuals: mean=%g stddev=%g' % (residuals.mean(), residuals.std()))
mpl.hist(residuals, bins=8)
mpl.xlabel('Residuals (mGal)')
mpl.ylabel('# of')

myv.figure()
myv.prisms(model, 'density', style='wireframe')
myv.prisms(vremove(0, 'density', mesh), 'density')
myv.axes(myv.outline(bounds), ranges=[i*0.001 for i in bounds], fmt='%.1f',
    nlabels=6)
myv.wall_bottom(bounds)
myv.wall_north(bounds)
myv.show()
```

```python
solver.fit()
mesh.addprop('density', solver.estimate_)
predicted = solver.predicted()

mpl.figure(figsize=(14, 5))
mpl.title("gz")
mpl.axis('scaled')
levels = mpl.contourf(yp, xp, gz, shape, 12)
mpl.colorbar()
mpl.contour(yp, xp, predicted, shape, levels, color='k')
mpl.xlabel('Horizontal coordinate y (km)')
mpl.ylabel('Horizontal coordinate x (km)')
mpl.m2km()

mpl.figure()
residuals = gz - predicted
mpl.title('Residuals: mean=%g stddev=%g' % (residuals.mean(), residuals.std()))
mpl.hist(residuals, bins=8)
mpl.xlabel('Residuals (mGal)')
mpl.ylabel('# of')

myv.figure()
myv.prisms(model, 'density', style='wireframe')
myv.prisms(vremove(0, 'density', mesh), 'density')
myv.axes(myv.outline(bounds), ranges=[i*0.001 for i in bounds], fmt='%.1f',
    nlabels=6)
myv.wall_bottom(bounds)
myv.wall_north(bounds)
myv.show()
```

```python
np.all(solver.estimate_ == np.array(estimate['density']))
```

```python
def harvester_run():
    #harvester.harvest([harvester.Gz(xp, yp, zp, gz), harvester.Gzz(xp, yp, zp, gzz)], seeds, mesh, compactness, threshold)
    harvester.harvest([harvester.Gz(xp, yp, zp, gz)], seeds, mesh, compactness, threshold)
%timeit -n 1 harvester_run()
```

```python
def blocky_run():
    #(Gravity(xp, yp, zp, gz, mesh) + Gravity(xp, yp, zp, gzz, mesh, field='gzz')).config(**args).fit()
    Gravity(xp, yp, zp, gz, mesh).config(**args).fit()
%timeit -n 1 blocky_run()
```
